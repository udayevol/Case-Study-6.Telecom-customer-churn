**Question 4: Is the company losing high-value customers? If so, how can they retain them?**

```python


# Define high-value threshold (e.g., top 25% in terms of total revenue)
high_value_threshold = churn['Total Revenue'].quantile(0.75)

# Identify high-value customers
high_value_customers = churn[churn['Total Revenue'] > high_value_threshold]

# Calculate churn rate among high-value customers
high_value_churned = high_value_customers[high_value_customers['Customer Status'] == 'Churned']
high_value_churned_rate = len(high_value_churned) / len(high_value_customers) * 100

print(f"High-Value Customer Churn Rate: {high_value_churned_rate:.2f}%")

# Analyze churn reasons for high-value customers
high_value_churn_reasons = high_value_churned['Churn Reason'].value_counts()


plt.figure(figsize=(12, 8))
high_value_churn_reasons.plot(kind='bar', color='orange')
plt.title('Churn Reasons for High-Value Customers')
plt.xlabel('Churn Reason')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Retention Strategies
# Example: If "Competitor Offer" is a major churn reason, consider strategies like loyalty programs, better offers, etc.
```
